```python
# Convert JSON strings to dictionaries
travel_cities["avg_temp_monthly"] = travel_cities["avg_temp_monthly"].apply(json.loads)

# Use json_normalize to flatten the JSON column
temp_monthly = pd.json_normalize(travel_cities["avg_temp_monthly"])
```

```python
# Concatenate the result with the original DataFrame
travel_cities = pd.concat([travel_cities.drop(columns=["avg_temp_monthly"]), temp_monthly], axis=1)
```
